```python
# Import Dependencies
import pandas as pd
```

```python
df1 = "201801-citibike-tripdata.csv"
df2 = "201802-citibike-tripdata.csv"
df3 = "201803-citibike-tripdata.csv"
df4 = "201804-citibike-tripdata.csv"
df5 = "201805-citibike-tripdata.csv"
df6 = "201806-citibike-tripdata.csv"
df7 = "201807-citibike-tripdata.csv"
df8 = "201808-citibike-tripdata.csv"
df9 = "201809-citibike-tripdata.csv"
df10 = "201810-citibike-tripdata.csv"
df11 = "201811-citibike-tripdata.csv"
df12 = "201812-citibike-tripdata.csv"
```

```python
df1 = pd.read_csv(df1)
df2 = pd.read_csv(df2)
df3 = pd.read_csv(df3)
df4 = pd.read_csv(df4)
df5 = pd.read_csv(df5)
df6 = pd.read_csv(df6)
df7 = pd.read_csv(df7)
df8 = pd.read_csv(df8)
df9 = pd.read_csv(df9)
df10 = pd.read_csv(df10)
df11 = pd.read_csv(df11)
df12 = pd.read_csv(df12)
```

```python
result = df1.append([df2, df3,df4, df5,df6, df7,df8, df9,df10, df11, df12])
```

```python
result.count()
```

```python
result= result[:10000000]
```

```python
result.head()
```

```python
result['StartDates'] = pd.to_datetime(result['starttime']).dt.strftime("%m/%d/%y")
result['StartTime'] = pd.to_datetime(result['starttime']).dt.time
result['StopDates'] = pd.to_datetime(result['stoptime']).dt.strftime("%m/%d/%y")
result['StopTime'] = pd.to_datetime(result['stoptime']).dt.time
```

```python
result.sort_values(by=['StartDates'])
```

```python
result.reset_index(drop=True, inplace=True)
```

```python
result.head()
```

```python
result.count()
```

```python
# Export to csv
result.to_csv("merged_data_2018.csv")
```
